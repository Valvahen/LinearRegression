```python
def compute_gradient(X,y,w,b):
    m,n = X.shape
    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range (m):
     err = (np.dot(X[i],w)+b)-y[i]
     for j in range(n):
        dj_dw[j]=dj_dw[j] + err * X[i,j]
     dj_db = dj_db +err  
    dj_dw = dj_dw/m
    dj_db = dj_db/m

    return dj_db,dj_dw 
```
